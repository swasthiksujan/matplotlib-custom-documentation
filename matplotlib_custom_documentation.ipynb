Getting started with `matplotlib`

```python

import matplotlib.pyplot as plt
import numpy as np
```

### Simple Plotting
---

We use the `plt.plot(<axes>)` funtions to plot the graph

To display the plotted graph we use the `plt.show()` function

```python
x = [2022, 2023, 2024, 2025, 2026]
y = [10, 20, 30, 40, 60]

plt.plot(x, y)
plt.show()
```

### Plot Customization
---

To customize our plot we can use multiple keyword arguments such as:
 - `marker` -> to control/change the marker style
     - `o` for circle
     - `*` for star
     - `v` for inverted triangle
     - `.` for dot

 - `markerfacecolor` or `mfc` can be used to control the color of the face of the marker
     - Syntax: `markerfacecolor="<color>"` -- Any color format such as rgb, hexadecimal can be used

 - `markeredgecolor` or `mec` controls the color of the edge of the marker

 - `markersize` or `ms` controls the size of the marker

 - `family="<fontname>"` controlls the font-family

 - `fontweight` can control the weight of the font. 

 - `linestyle` can control the style of the line
     - `solid` is the default and provides a solid line ( _________ )
     - `dashed` is ( ----- )
     - `dotted` is ( .... )
     - `dashdot` is ( _._._. )

 - `linewitdh` controls the width of the line 
 
 - `color` controls the colour of the line in the graph


```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 20]

plt.plot(x, y, marker="*", 
         markerfacecolor="red",
         markeredgecolor="black",
         markersize=10,
         linestyle="dotted",
         linewidth=3,
         color="violet")

plt.show()
```

### Applying Styling to multiple lines in the same plot
---

To apply multiple styles to different lines in the plot, we can make a dictionary with the keyword arguments we used earlier. When we need to access these values, we can unpack the dictionary as needed.

To unpack a dictionary, we use `**<dictionary_name>`

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 20]
y1 = [5, 10, 15, 20, 7]

line_Style = dict(marker="o", 
         markerfacecolor="cyan",
         markeredgecolor="darkblue",
         markersize=10,
         linestyle="dashed",
         linewidth=3)

plt.plot(x, y, **line_Style, color="red")
plt.plot(x, y1, **line_Style, color="green")

plt.show()
```

### Adding Title, Labels Grids and Ticks
---

We can add title to a plot using the `plt.title()` function

To add a label, we have to add labels to x and y axis separately. To do say we make use of
 - `plt.xlabel()`
 - `plt.ylabel()`

These titles and labels can be customized using certain keyword arguments

To add a grid to the plot, we use `plt.grid()` function. We can customise the axis of the grid and some other details too. To do so, we specify the axis and the arguments
 - `plt.grid(axis=x)`
 - `plt.grid(axis=y)`

 To force the x or y axis to only show the numbers when the value is near, we can use the `plt.xticks()` and `plt.yticks()` functions.
  - Pass in the list/the dataset you have for each axis as arguments. 

To customize the ticks, we have another function called `plt.tick_params()`

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 20]
y1 = [5, 10, 15, 20, 7]

line_Style = dict(marker="o", 
         markerfacecolor="cyan",
         markeredgecolor="darkblue",
         markersize=10,
         linestyle="dashed",
         linewidth=3)

text_Style = dict(family="Segoe UI",
                  fontweight="bold",
                  color="darkblue")

plt.title("Plot Title", **text_Style, fontsize=20)
plt.xlabel("Year", **text_Style, fontsize=12)
plt.ylabel("Random Numbers", **text_Style, fontsize=12)

plt.grid(linestyle="dotted", color="lightgrey", linewidth=2)

plt.xticks(x)
plt.yticks(y)
plt.tick_params(axis = "both", 
                colors="darkgreen", 
                labelsize=10,
                pad=10)

plt.plot(x, y, **line_Style, color="red")
plt.plot(x, y1, **line_Style, color="green")

plt.show()
```

## Bar Charts

Use the `plt.bar(<xaxis>, <yaxis>)`

### Pie Chart
---

Good for visualizing parts of whole, self explanatory ig

We use the `plt.pie()` function to do so. Here are some of the arguments that can be passed into the function to customise the pie chart

 - `labels=<any_iterable_dataType>` -> Assigns labels to each of the slices

 - `autopct="%1.<number_of_decimal_places>f%%` -> To make the slices display the percentage. The formatting for percentage display is wierd, but easy once you are used to it

 - `colors=[<list_of_colors>]` -> Applies custom colors to the slices

 - `explode=[<list of distances>]` -> Separates the slices of the pie by a certain distance as set by the user in the form of a list. This is used to highlight or explicitly point to a certain slice. Number of slices and the length of the distance list must match for this to work.

 - `shadow=<bool>` -> Applies the shadow effect
 
 - `startangle=<angle_in_degrees>` -> Adjusts the rotation of the pie chart. 


```python
categories = np.array(["freshmen", "sophomores", "juniors", "seniors"])
population = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "green", "gold"]

plt.title("College Stats")
plt.pie(population, labels=categories,
        autopct="%1.2f%%",
        colors=colors,
        explode=[0, 0, 0, 0.1],
        shadow=True,
        startangle=180)


plt.show()
```

### Scatter Graphs
---

Used to display corrolations between two variables

Here are some of the keyword arguments that can be passed into the `plt.scatter(<x_axis>, <y_axis>)` function
 - `color` -> changes the color of the dots

 - `alpha` -> adjusts the transparency 

 - `s` -> size of the dots

 - `label` -> specifies the label for displaying legends

To display Legends we can make use of the `plt.legends()` function.

```python
hours_studied = np.array([0, 1, 3, 4, 1, 3, 2, 5, 7, 10, 10])
grades = np.array([55, 60, 65, 68, 70, 76, 82, 85, 87, 95, 96])

plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Hours Studies and Grades - A Correlation Graph")

plt.scatter(np.sort(hours_studied), np.sort(grades),
            color="black",
            alpha=0.5,
            s = 100,
            label="Class A") # sorted them here to get a positive correlation hehe
plt.legend()
plt.show()
```
